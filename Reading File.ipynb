# Read a csv file
## First line represents Header
## Second line data type
## Rest of the lines information
## Sort them by metro population

```python
from functools import namedtuple
def cast_row(data_type, data_value):
    output = []
    for type_, value in zip(data_type, data_value):
        if type_ == 'STRING':
            value = str(value)
        elif type_ == 'INT':
            value = int(value)
        else:
            raise TypeError('Unsupported Data Type.')
        output.append(value)
    return output
```

```python
with open('population cities.csv') as file:
    file_iter = iter(file)
    header = next(file_iter).strip().split(',')
    Population = namedtuple('Population',header)
    data_type_row = next(file_iter).strip().split(',')
    out = [Population(*cast_row(data_type_row, line.strip().split(','))) 
           for line in file_iter]
    out.sort(key=lambda x:x.Metro_Population, reverse = True)
    for row in out:
        print(row)
```
