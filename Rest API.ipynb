# Importing Libraries

```python
from urllib import request
import json
```

# Calling a basic GET - RestAPI

```python
url = 'http://dummy.restapiexample.com/api/v1/employees'
res = request.urlopen(url)
```

```python
json_response = json.loads(res.read())
```

```python
json_response
```

# Sorting the response result by name

```python
sorted_response = sorted(json_response, key = lambda x: x['employee_name'], reverse = True)
```

```python
sorted_response
```

# Creating a tuple

```python
from collections import namedtuple
```

```python
header = json_response[0].keys()
```

```python
Employee = namedtuple('Employee',header)
```

```python
emp_list = [Employee(**employee) for employee in json_response]
```

```python
for emp in emp_list:
    print(emp)
```

```python
emp_generator = (Employee(**employee) for employee in json_response)
```

```python
from itertools import cycle, islice
emp_generator_cycle = cycle(emp_generator)
```
